```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações
S0 = 100
r = 0.04
q = 0.03
sigma = 0.25
T = 0.5
participation = 0.4
capital = 100000
n_sim = 100000

np.random.seed(1)
Z = np.random.randn(n_sim)

# Simulação do preço do índice ao fim do período
ST = S0 * np.exp((r - q - 0.5 * sigma ** 2) * T + sigma * np.sqrt(T) * Z)
R = ST / S0 - 1  # retorno percentual

# Payoff do produto
payoffs = np.maximum(0, participation * R) * capital * np.exp(-T*r)
expected_payoff = np.mean(payoffs)

# Retorno no ativo livre de risco
risk_free_return = capital * (np.exp(r * T) - 1)

# Probabilidade de payoff positivo
prob_gain = np.mean(payoffs > 0) * 100

# -------------------- Gráficos --------------------

sns.set(style="whitegrid")

# 1) Distribuição dos retornos do índice
plt.figure(figsize=(8,4))
sns.histplot(R, bins=100, kde=True)
plt.title('Distribuição dos Retornos do Índice em 6 meses')
plt.xlabel('Retorno (%)')
plt.ylabel('Frequência')
plt.show()

# -------------------- Resultados Numéricos --------------------
print(f"Payoff esperado do produto: ${expected_payoff:.2f}")
print(f"Retorno no ativo livre de risco: ${risk_free_return:.2f}")
print(f"Probabilidade de ter ganho (payoff > 0): {prob_gain:.2f}%")
```
